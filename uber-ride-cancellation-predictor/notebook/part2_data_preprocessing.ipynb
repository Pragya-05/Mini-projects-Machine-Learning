```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/uber_rides.csv")

print(df.shape)
df.head()
```

```python
df = df[
    df["Booking Status"].isin(
        [
            "Completed",
            "Cancelled by Customer",
            "Cancelled by Driver",
            "No Driver Found"
        ]
    )
]

df["is_cancelled"] = np.where(
    df["Booking Status"] == "Completed",
    0,
    1
)

df["is_cancelled"].value_counts()
```

```python
df["hour_of_day"] = df["hour"]

df[["hour", "hour_of_day"]].head()
```

```python
df[["Avg VTAT", "Avg CTAT"]].isnull().sum()
```

```python
df["Avg VTAT"] = df["Avg VTAT"].fillna(
    df["Avg VTAT"].median()
)

df["Avg CTAT"] = df["Avg CTAT"].fillna(
    df["Avg CTAT"].median()
)
```

```python
df[["Avg VTAT", "Avg CTAT"]].isnull().sum()
```

```python
columns_to_drop = [
    "Booking Status",
    "Ride Distance",
    "Booking Value",
    "Driver Ratings",
    "Customer Rating",
    "Booking ID",
    "Customer ID",
    "Date",
    "Time"
]

df = df.drop(columns=columns_to_drop)
```

```python
df.columns.tolist()
```

```python
print(df.shape)
print(df.columns.tolist())
```

```python
columns_to_drop = [
    "Cancelled Rides by Customer",
    "Reason for cancelling by Customer",
    "Cancelled Rides by Driver",
    "Driver Cancellation Reason",
    "Incomplete Rides",
    "Incomplete Rides Reason"
]

df = df.drop(columns=columns_to_drop)

print(df.columns.tolist())
```

```python
df.isnull().sum()
```

```python
df["Payment Method"] = df["Payment Method"].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=["hour"])
```

```python
df = df.drop(columns=["datetime", "ride_date"])
```

```python
print(df.shape)

print(df.columns.tolist())
```

```python
df.to_csv(
    "../data/clean_uber_rides_2024.csv",
    index=False
)

print("Clean dataset saved successfully!")
```
